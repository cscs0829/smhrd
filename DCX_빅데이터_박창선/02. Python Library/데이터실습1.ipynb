```python
# 그룹화를 이용한 통계 분석 : groupby()
```

```python
# 여러 컬럼 인덱싱 : 대괄호로 한번 더 감싸주기!
hr[['SALARY', 'EDU_LEVEL']].groupby(by = 'EDU_LEVEL')
```

```python
for i, df in hr[['SALARY','EDU_LEVEL']].groupby(by = 'EDU_LEVEL'):
    display(i) # 학위
    display(df) # 데이터프레임
```

```python
hr[['SALARY','EDU_LEVEL']].groupby(by = 'EDU_LEVEL').agg(['mean','sum','max'])
```

```python
# 그룹화를 이요한 통계 분석 : groupby()
```

```python
# 여러 컬럼 인덱싱 : 대괄호로 한번 더 감싸부기!
hr[['SALARY','EDU_LEVEL']].groupby()

# TypeError : you have to supply one of 'by' and 'level'
# 'by' or 'level'  인자값을 줘야함!
# 'by' : 컬럼이름, 'level' = 인덱스 번호 or 인덱스 이름
```

```python
# 그룹화 후 여러 통계값 확인하기
# 교육수준, 성별에 따른 평균 연봉 확인
hr[['SALARY','EDU_LEVEL','GENDER']].groupby(by = ['EDU_LEVEL', 'GENDER']).agg(['mean','max','min','median'])
```

```python
# 그룹화 후 컬럼별로 다른 연산 적용
# 국적에 따른 평균 성과평과점수와 최대 연봉 확인하기
# 'NATION', 'PERFORMANCE_SCORE', 'SALARY' 
hr.groupby(by = 'NATION').agg({'PERFORMANCE_SCORE':'mean', 'SALARY':'max'})
```

### 상관계수 분석
- 두 변수 사이의 관계를 수치화해서 보는 방법


```python
# 직원 성과평가점수와 연봉의 상관 계수 확인 : corr()
hr[['PERFORMANCE_SCORE', 'SALARY']].corr()
```
