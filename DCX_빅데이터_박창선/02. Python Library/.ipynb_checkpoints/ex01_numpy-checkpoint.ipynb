### BMI

```python
#  ()
# txt 
# delimiter = ''
# np.loadtxt('', delimiter = ?)
data = np.loadtxt('data/height_weight.txt', delimiter=',')
data
```

```python
#    ->  

# data  
print(data.ndim)
# data 
print(data.dtype)
# data ,  
print(data.shape)
```

```python
# 1.  (h)   
h = np.array(data[0])
h
```

```python
# 2.  (w)   
w = np.array(data[1])
w
```

```python
# 3.     (1m = 100cm)
h_100 = h*0.01
h_100
```

```python
# 4.    bmi  
# bmi   :  /  * 
bmi = w/(h_100*h_100)
bmi
```

```python
# 5.    (  )
bmi[bmi >= 23]
```

```python
bmi[(23<=bmi)&(bmi<25)]
```
