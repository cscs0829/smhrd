```python
# for문을 사용해 파일 불러오기

# 빈 리스트 생성해서 df 추가하기
df_list = []
for file in os.listdir('./월별급여'):
    df = pd.read_csv('월별급여/'+file, encoding = 'euc-kr')
    df_list.append(df)
    display(df)
```

```python
# 병합한 데이터프레임을 csv파일로 저장하기
total_df.to_csv('./월별급여/total_pay_info.csv', index = False)
```

#### org-info 데이터에 각 조직별로 인원 수, 평균 나이, 평균 근무 년수, 평균 연봉 추가

```python
# org-info.csv 파일 불러오기
org_info = pd.read_csv('org-info.csv', encoding = 'euc-kr', index_col = 'ORG_CD')
org_info.head(3)
```

```python
# 조직별로 인원 수, 평균 나이, 평균 근무 년수, 평균 연봉 -> hr-info.csv 파일에 관련 내용 포함
hr_info = pd.read_csv('hr-info.csv', encoding = 'euc-kr')
hr_info
```

```python
# hr_info에서 groupby를 이용해서 조직별 인원 수, 평균 나이, 평균 근무 년수, 평균 연봉에 대한 데이터프레임 생성
org_agg = hr_info.groupby(by = 'ORG_CD').agg({'EID' : 'count',
                                   'AGE' : 'mean',
                                   'SERVICE_YEAR' : 'mean',
                                   'SALARY' : 'mean'})
org_agg
```

```python
hr_info.columns
```

```python
# org_info랑 추출한 데이터(org_agg)를 병합
# 행의 방향으로 넣는게 아닌 열의 방향
org_info.head(2)
```

```python
pd.concat([org_info, org_agg], axis = 1)
```

- merge()
    - 데이터베이스의 조인 개념을 떠올리면 좋다(inner, outer, left, right)
    - 공통된 컬럼을 기준으로 양쪽의 데이터를 병합할때 활용


```python
# hr_info와 급여 정보(total_pay_info) 병합
```

```python
# 급여 정보 파일 불러오기
hr_pay_info = pd.read_csv('./월별급여/total_pay_info.csv')
hr_pay_info
```

```python
hr_info.columns
```

```python
print(hr_info.shape)
print(hr_pay_info.shape)
```

```python
# 둘 다 가지고 있는 'EID' 컬럼을 중심으로 merge 해보기
pd.merge(hr_info, hr_pay_info, how = 'inner', on = 'EID')
```

```python
# hr_info 기반으로 merge
merge_df = pd.merge(hr_info, hr_pay_info, how = 'left', on = 'EID')
merge_df
```

```python
# 급여 내역을 기반으로 merge
pd.merge(hr_info, hr_pay_info, how = 'right', on = 'EID')
```

```python
merge_df.shape
```

```python
# 지급 내역이 없는 인원 확인하기
no_money = merge_df[merge_df['PAYMENT_DY'].isnull()]
no_money
```

```python
# 지급내역이 없는 사람들이 언제 퇴사를 했는지 알아보자
# hr_info의 'RETIRE_DATE' 컬럼 활용
pd.merge(no_money, hr_info)[['ENAME','RETIRE_DATE']]
```

```python
def skill_point(row):
    skill_point = 0
    for skill in row['SKILLS']:
        
        if 'Java' in skill or 'Python' in skill:
            skill_point += 1
    return skill_point
```

```python
def skill(row):
    point = 0
    if 'Java' in row['SKILLS'] or 'Python' in row['SKILLS']:
        point += 1
    if 'Data Analysis' in row['SKILLS']:
        point += 2
    return point
```

```python
hr_info['SKILLS']
```

```python
hr_info.apply(skill, axis = 1)
```
