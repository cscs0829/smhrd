### 영재팀

```python
# / : 절대경로, 루트부터 시작
# ./ : 절대값, 상대경로, 현재 위치
# ../ : 상대경로, 상위(부모)

with open("./data/영재팀(와인).pkl", "rb") as f :
    영재팀_와인data = pickle.load(f)
```

```python
# 데이터프레임으로 변환!
영재팀_와인df = pd.DataFrame.from_dict(
    영재팀_와인data,
    orient = "index"
)
```

```python
# 요약 정보 확인 (기술통계치 확인)
영재팀_와인df.describe().T
# 전치 : 행과 열을 바꿈
```

### 선웅팀

```python
with open("./data/선웅팀(의류데이터)/user_data.pkl", "rb") as f : 
    선웅팀_user_data = pickle.load(f)

with open("./data/선웅팀(의류데이터)/review_data.pkl", "rb") as f : 
    선웅팀_review_data = pickle.load(f)

with open("./data/선웅팀(의류데이터)/product_data.pkl", "rb") as f : 
    선웅팀_product_data = pickle.load(f)

with open("./data/선웅팀(의류데이터)/style_data.pkl", "rb") as f : 
    선웅팀_style_data = pickle.load(f)
```

```python
선웅팀_product_df = pd.DataFrame.from_dict(선웅팀_product_data, orient = "index")
선웅팀_review_df = pd.DataFrame.from_dict(선웅팀_review_data, orient = "index")
선웅팀_style_df = pd.DataFrame.from_dict(선웅팀_style_data, orient = "index")
선웅팀_user_df = pd.DataFrame.from_dict(선웅팀_user_data, orient = "index")
```

```python
display(선웅팀_product_df)
display(선웅팀_review_df)
display(선웅팀_style_df)
display(선웅팀_user_df)
```

```python
# 모양 확인
선웅팀_product_df.shape, 선웅팀_review_df.shape, 선웅팀_style_df.shape, 선웅팀_user_df.shape
```

```python
# 조인도 가능!
pd.merge(선웅팀_user_df, 선웅팀_style_df, on = "user_id")
```

```python
# 조인도 가능!
선웅팀_user_into_df = pd.merge(선웅팀_user_df, 선웅팀_style_df, on = "user_id")
선웅팀_user_into_df.head()
```
