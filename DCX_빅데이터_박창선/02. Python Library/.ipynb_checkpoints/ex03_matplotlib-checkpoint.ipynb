### 데이터 시각화 라이브러리(matplotlib)
- 방대한 양의 자료를 한 눈에 볼 수 있도록 도표나 차트 등으로 정리
- 시각화를 통해 데이터의 특징을 쉽게 파악 가능
- 분석 결과를 상대방에게 효과적으로 전달 가능
- matplotlib을 이용한 시각화 : 한땀한땀 직접 그려볼 수 있다. 커스터마이징에 용이
- pandas, seaborn을 이용한 시각화 : 손쉽게 차트를 그려볼 수 있다

```python
import pandas as pd
```

1. line chart(선 그래프)
- hr-info 데이터에서 연도별 입사자 추세 시각화

```python
# hr-info.csv 파일 불러오기
# 날짜 타임으로 변환할 컬럼 설정 속성 : parse_dates
hr = pd.read_csv('./data/hr-info.csv', encoding = 'euc-kr', parse_dates=['HIRE_DATE'])
hr.head()
```

```python
hr.info()
```

```python
# 입사년도 추출하기
hr['HIRE_DATE'].dt.year
```

```python
# 새로운 컬럼 'HIRE_YEAR' 추가
hr['HIRE_YEAR'] = hr['HIRE_DATE'].dt.year
hr.head()
```

```python
# 입사년도별 인원 수 파악
hr['HIRE_YEAR'].value_counts().sort_index()
```

```python
# 위의 데이터를 년도별로 오름차순으로 정렬하기
hr['HIRE_YEAR'].value_counts().sort_index()
```

```python
# 위의 데이터 선그래프로 표현하기 
# kind = 그래프 유형
# xlabel, ylabel : x축 이름, y축 이름
# grid = True : 모눈종이
hr['HIRE_YEAR'].value_counts().sort_index().plot(kind ='line',
                                                 xlabel = 'HIRE_DATE',
                                                ylabel='HIRE_COUNT',
                                                grid = True)
```

### 2.bar chart(막대 그래프)
- 학위별 평균 연봉 비교 시각화

```python
hr.columns
```

```python
# 데이터 그룹화 후 통계함수 사용
# 'EDU_LEVEL', 'SALARY'
hr[['EDU_LEVEL','SALARY']].groupby(by = 'EDU_LEVEL').agg(['mean', 'sum', 'min'])
```

```python
# 차트에 한글을 출력하려면 인코딩 작업 필요
from matplotlib import rc
rc('font', family = 'Malgun Gothic')
```

```python
# 위 데이터 시각화 하기 
hr[['EDU_LEVEL','SALARY']].groupby(by = 'EDU_LEVEL').agg('mean').plot(kind = 'barh',
                                                                     xlabel = '학위',
                                                                     ylabel = '평균 연봉',
                                                                     title = '학위별 평균 연봉')
```

```python
# 학위, 성별을 고려한 평균 연봉 시각화
# round(n) : 소수점 n째자리 까지 가져오기
hr[['EDU_LEVEL', 'SALARY', 'GENDER']].groupby(by = ['EDU_LEVEL', 'GENDER']).mean().round(1).plot(kind = 'bar')
```

```python
# 데이터 재구조화
sal_mean = hr[['EDU_LEVEL', 'SALARY', 'GENDER']].groupby(by = ['EDU_LEVEL', 'GENDER']).mean().round(1)
```

```python
sal_mean2 = sal_mean.reset_index().pivot(index = 'EDU_LEVEL', columns = 'GENDER', values = 'SALARY')
sal_mean2
```

```python
sal_mean2.plot(kind = 'barh', colormap = 'Set3', subplots = True)
```

```python
# 부서별 평균 연봉 비교하기(시각화)
```

```python
# hr-info.csv 파일 불러오기
hr_info = pd.read_csv('./data/hr-info.csv',  encoding = 'euc-kr')
hr_info
```

```python
# org-info.csv 파일 불러오기
org_info = pd.read_csv('./data/org-info.csv',  encoding = 'euc-kr')
org_info
```

```python
# 조직정보 , 구성원 정보 merge
# ORG_CD 중심으로 merge
merge_df = pd.merge(org_info, hr_info, how = 'inner', on = 'ORG_CD')
merge_df
```

```python
merge_df.columns
```

```python
# 부서별 평균 연봉 구하기 -> 시각화하기
# 'SALARY', 'KOR_TEAM'
merge_df[['KOR_TEAM','SALARY']].groupby(by = 'KOR_TEAM').agg('mean')
```

```python
merge_df[['KOR_TEAM', 'SALARY']].groupby(by = ['KOR_TEAM']).mean().plot(kind = 'bar', figsize= (20,3), ylim = (4000,7000))

# figsize = 그림 크기 조절하는 속성, inch 단위
# ylin = y축 limit 제한 속성, 축의 최소, 최대값 설정
```

3. pie chart(원그래프)
    - 부서별 인원 비율 시각화

```python
merge_df['KOR_DIVISION'].value_counts().plot(kind = 'pie',
                                            autopct = '%.1f%%', # 원그래프 안에 숫자 표시
                                            startangle = 90, # 원 그래프의 첫 데이터 시작 위치 설정, 90 : 12시를 기준으로 시작
                                            counterclock = False, # False : 시계방향, True : 반시계방향
                                            ylabel = ''
                                            )
```

4. scatter(분포도, 산점도)
- 나이에 따른 연봉 변화 시각화

```python
hr.plot(kind = 'scatter', x='AGE', y='SALARY', color='red')
```

### Seaborn 라이브러리를 이용한 그래프 시각화
- 손쉽게 그래프를 표현할 수 있다
- 팔레트가 다양해서 화려하고 다양한 형태의 그래프를 활용할 수 있다

```python
import seaborn as sns
```

1. 히스토그램
- 성별에 따른 연봉 시각화

```python
sns.histplot(data = hr, # 사용할 데이터
             x = 'SALARY',
             hue = 'GENDER',
             multiple = 'stack'
            )
```

2. 버블차트
- 나이와 근속년수에 따른 직급별 연봉 비교 시각화

```python
hr.columns
```

```python
sns.relplot(hr, x='AGE', y = 'SERVICE_YEAR', hue = 'POS_CD', size='SALARY', sizes=(40,1200), alpha = 0.5, height = 6)
```

3. 히트맵
- 상관계수 시각화

```python
# 성과점수, 연봉, 나이, 근속년수, 훈련시간의 상관계수 시각화
hr_corr = hr[['PERFORMANCE_SCORE', 'SALARY', 'AGE', 'SERVICE_YEAR','TRAINING_HOURS']].corr()
hr_corr
```

```python
sns.heatmap(hr_corr, annot = True)
# annot = True : 히트맵 안에 숫자 표현
```

### matplotlib 라이브러리를 활용한 시각화

```python
import matplotlib.pyplot as plt
```

```python
# 선그래프 그리기 
x = [1,2,3,4]
y = [2,4,6,8]

plt.plot(x,y, marker = 'o')

# marker = 값의 위치를 표시
# color(c) : 선의 색깔
# lineStyle(ls) : 선의 스타일 설정
# linewidth(lw) : 선의 두께 설정
# MarkerSize(ms) : 마커 사이즈 변경
# MarkerFaceColor(mfc) : 마커 색 변경
# MarkerEdgeColor(mec) : 마커 선 색 변경
```

```python
fig, ax = plt.subplots(figsize=(6, 6))


# 결과 출력
plt.show()
```
