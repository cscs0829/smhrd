```python
import random

# 각 스타일에 따른 의류 아이템 목록 정의
item_options = {
    "캐주얼": ["청바지", "티셔츠", "후드티", "스니커즈"],
    "포멀": ["정장 바지", "셔츠", "블레이저", "구두"],
    "스트릿": ["조거 팬츠", "그래픽 티셔츠", "캡", "운동화"],
}

# 나이대, 성별, 스타일, 계절, 색상 목록 정의
age_groups = ["20대", "30대", "40대"]
genders = ["남성", "여성"]
styles = ["캐주얼", "포멀", "스트릿"]
seasons = ["봄", "여름", "가을", "겨울"]
colors = ["검정", "흰색", "빨강", "파랑", "초록"]

# 메타데이터 생성
product_metadata = []

for age in age_groups:
    for gender in genders:
        for style in styles:
            for season in seasons:
                for color in colors:
                    # 랜덤하게 아이템 선택
                    item = random.choice(item_options[style])
                    product_metadata.append({
                        "상품명": f"{age} {gender} {item} {season} 아이템",
                        "나이대": age,
                        "성별": gender,
                        "색상": color,
                        "패션스타일": style,
                        "계절": season
                    })

# 사용자 입력 받기
print("의류 추천 시스템에 오신 것을 환영합니다!")

def get_age_group(age):
    if 20 <= age < 30:
        return "20대"
    elif 30 <= age < 40:
        return "30대"
    elif 40 <= age < 50:
        return "40대"
    else:
        return None

user_profile = {
    "나이대": get_age_group(int(input("나이를 입력하세요: "))),
    "성별": input("성별을 입력하세요 (남성/여성): "),
    "선호 색상": input("선호 색상을 입력하세요 (예: 검정, 흰색, 빨강 등): "),
    "패션 스타일": input("패션 스타일을 입력하세요 (캐주얼, 포멀, 스트릿 등): "),
    "계절": input("현재 계절을 입력하세요 (봄, 여름, 가을, 겨울): ")
}

# 추천 로직
def recommend_clothes(profile, products):
    recommendations = []
    for product in products:
        if (
            product["나이대"] == profile["나이대"] and
            product["성별"] == profile["성별"] and
            product["색상"] == profile["선호 색상"] and
            product["패션스타일"] == profile["패션 스타일"] and
            product["계절"] == profile["계절"]
        ):
            recommendations.append(product)
    
    return recommendations

# 의류 추천 결과 출력
recommended_clothes = recommend_clothes(user_profile, product_metadata)
if recommended_clothes:
    print("\n추천 의류 목록:")
    for item in recommended_clothes:
        print(f"- 상품명: {item['상품명']}")
else:
    print("\n추천할 의류가 없습니다. 조건을 변경해보세요!")

```

```python
import pickle

# 파일에서 데이터 불러오기
with open("./실습자료.pkl", "rb") as f:
    dict2 = pickle.load(f)

# 데이터 가져오기
data = dict2["data"]

# 사용자 입력 받기
print("상품 추천 시스템")

category = input("카테고리를 입력하세요 (상의/하의/아우터/신발 등): ")
color = input("색상을 입력하세요 (예: 블루, 베이지 등): ")
season = input("계절을 입력하세요 (봄/여름/가을/겨울): ")
style = input("스타일을 입력하세요 (캐주얼/포멀 등): ")
age_group = input("나이대를 입력하세요 (20대/30대 등): ")

# 추천 로직
def find_products(data, category, color, season, style, age_group):
    recommendations = []
    for item in data:
        if (item["카테고리"] == category and
            item["색상"] == color and
            item["계절"] == season and
            item["스타일"] == style and
            item["나이대"] == age_group):
            recommendations.append(item["상품명"])
    return recommendations

# 상품명 찾기
recommended_products = find_products(data, category, color, season, style, age_group)

# 결과 출력
if recommended_products:
    print("\n추천 상품:")
    for product in recommended_products:
        print(f"- {product}")
else:
    print("\n추천할 상품이 없습니다. 조건을 변경해보세요.")
```
