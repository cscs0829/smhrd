# 네이버 뉴스데이터 수집

## 실습 프로세스
1. 네이버 뉴스 페이지 접속
2. IT/과학 카테고리 접속
3. 헤드라인 뉴스 접속
4. 뉴스의 제목, 내용 가져오기
- URL주소로 뉴스페이지 요청(requests)
- 응답받은  웹문서(res.text)를 bs를 활용하여 변환
- select_one()를 활용해서 제목, 내용 수집


```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
news_url = "https://n.news.naver.com/mnews/article/009/0005418850"
res = req.get(news_url)
res.text
```

# 복습() CSS 선택자

1. 태그선택자: 태그 이름으로 요소를 선택
2. 클래스선택자: "."으로 요소를 선택
3. 아이디선택자: "#"으로 요소를 선택
4. 계층선택자
- 자식선택자: "A > b" 현태로 요소를 선택 (A:부모, B:자식)
- 자손선택자: "A B" 형태로 요소를 선택 (A:부모, B:자손)
- 인접형제선택자 "A + B" 형태로 요소를 선택 (A:형제1, B:형제2)
- 형제선택자: "A ~ B"형태로 요소를 선택 (A:형제1, B:형제2~형제n)

```python
html = bs(res.text,"lxml")
```

```python
html.select_one("#title_area > span").text
```

# 여러 페이지의 내용을 수집해야 할 경우 
- 눈에 보이지 않는 데이터 수집 필요 -> a태그의 href속성값

## 데이터 수집 순서
1. 뉴스 페이지의 URL주소 수집 및 리스트 저장
2. 리스트와 반복문을 활용해서 각 뉴스 페이지 문서를 요청
3. 각 뉴스 페이지 문서를 변환하는 작업
4. 변환된 문서 내 뉴스 제목과 내용 수집

```python
news_section_url = "https://news.naver.com/section/105"

res =req.get(news_section_url)

html = bs(res.text, "lxml")
```

```python
news_list = html.select("a.sa_text_title")
news_list
```

```python
news_list[0].text.strip()
```

```python
# get(태그의 속성이름): HTML요소의 속성값을 문자열로 반환
news_list[0].get("href")
```

## 실습) news_list에 저장된 모든 a요소의 href속성값을 리스트로 저장해보기!

```python
href_list = []

for i in news_list:
    href_list.append(i.get("href"))

href_list
```

```python
href_list = []

for i in range(len(news_list)):
    href_list.append(news_list[i].get("href"))

href_list
```

```python
# 2.
for href in href_list:
    res = req.get(href)

# 3.
    html = bs(res.text, "lxml")

# 4.
    tit = html.select_one("#title_area > span").text
    content = html.select_one("#dic_area").text.strip()

    print("뉴스제목:", tit)
    print("내용:", content)
    print("="*50)
```
