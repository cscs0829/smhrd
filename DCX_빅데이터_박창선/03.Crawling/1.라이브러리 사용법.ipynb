# requests
- 웹 사이트의 정보를 요청/응답 받기 위한 라이브러리

```python
import requests as req
```

```python
# 요청할 웹 사이트의 URL 주소를 변수에 저장
url = "https://www.naver.com"


# 200의 의미 : HTTP 상태코드의 한 종류로, 성공했다는 의미로 해석
# 400: 클라이언트 오류
# 500: 서버 오류
req.get(url)
```

```python
res = req.get(url)

#웹 사이트의 페이지 소스코드 접근
res.text
```

# BeautifulSoup
- 문자열인 웹 문서를 python으로 변환해주는 라이브러리
- 변환된 데이터를 통해 데이터 추출 가능

```python
from bs4 import BeautifulSoup as bs
```

```python
weather_url ="https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&ssc=tab.nx.all&query=%EB%82%A0%EC%94%A8&oquery=%EB%A7%88%EB%9D%BC%ED%86%A4&tqi=i2Cb6sqpts0sskrJ3gwsssssshK-042468"

res = req.get(weather_url)

res.text
```

```python
bs(res.text, "lxml")
```

```python
type(res.text), type(bs(res.text, "lxml"))
```

```python
# bs(변환할 데이터, 변환방식)
# - 변환할 데이터 : res.text
# - 변환바익 : Lxml > html.parser > html5lib
html = bs(res.text, "lxml")
```

```python
# select_one(CSS선택자): 조건에 해당하는 HTML요소 하나를 반환
# 요소를 찾을 때 위에서부터 아래로 내려오는 방식
html.select_one("title")
```

```python
# text속성 : HTML요소의 콘텐츠를 문자열로 반환
html.select_one("title").text
```
