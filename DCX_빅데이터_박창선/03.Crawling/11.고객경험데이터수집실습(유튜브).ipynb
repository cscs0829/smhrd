### 실습) 각 동영상 내 작성된 댓글 수집하기 
1. 유튜브 검색 결과 페이지에서 "동영상"탭 클릭
2. 동영상 링크를 수집한 후 빈 리스트에 저장
3. 반복문을 이용하여 수집한 링크로 페이지 요청
4. 페이지 영상 정지 및 스크롤 내리기 동작
5. 타임스탬프 삭제
6. 댓글 데이터 수집 후 텍스트 파일로 저장

```python
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from tqdm import tqdm

import time
import re

def preprocess_sentence_kr(w):
  w = w.strip()
  w = re.sub(r"[^0-9가-힣?.!,¿]+", " ", w)
  w = w.strip()
  return w
```

```python
keyword = '음식물%20처리기%20사용%20후기'
url = f'https://www.youtube.com/results?search_query={keyword}'
```

```python
driver = wb.Chrome()
```

```python
driver.get(yt_url)
```

```python
# 동영상 탭 클릭
# :nth-child(요소의 위치)
tab = driver.find_element(By.CSS_SELECTOR, '#chips > .style-scope.yt-chip-cloud-renderer:nth-child(3)')
tab.click()
```

```python
# 동영상 링크 수집 및 저장
videoTitles = driver.find_elements(By.ID, 'video-title')
len(videoTitles)

href_list = []

for title in videoTitles:
    href_list.append(title.get_attribute('href'))

href_list
```

```python
# 반복문을 이용하여 수집한 링크에 대한 페이지 요청
f = open('음식물 처리기 사용 후기_유튜브댓글수집.txt','w')

for i in tqdm(range(len(href_list))):
    driver.get(href_list[i])

    driver.implicitly_wait(3)

    play_btn = driver.find_element(By.CLASS_NAME, 'ytp-play-button')
    play_btn.click()
    #스크롤 동작
    old_height = driver.execute_script("return document.documentElement.scrollHeight")

    while True:
        body = driver.find_element(By.TAG_NAME,'body')
        body.send_keys(Keys.END)
        time.sleep(1)
        
        new_height = driver.execute_script("return document.documentElement.scrollHeight")
    
        if old_height == new_height:
            break
    
        old_height = new_height

    # 타임스탬프 요소 삭제 작업
    driver.execute_script('''
        const timeStamps = document.querySelectorAll("#content-text a.yt-core-attributed-string__link");
    
        for(let i=0; i<timeStamps.length; i++){
            timeStamps[i].remove();
        }
    ''')
    
    #댓글 수집
    comments = driver.find_elements(By.CSS_SELECTOR, '#content-text > span')

    for i in range(1, len(comments)):
        comment = preprocess_sentence_kr(comments[i].text)
        f.write(comment)
    
    time.sleep(1)
f.close()
```
