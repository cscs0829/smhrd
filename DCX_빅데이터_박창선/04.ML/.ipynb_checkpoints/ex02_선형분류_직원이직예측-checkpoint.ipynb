### 문제 정의
- BX : 최근 이직시장이 활발하게 성장함에 따라 직원들도 줄줄이 퇴사행렬을 이어가고 있다. 경영진은 핵심 인력 유출을 최대한 출이고 오랜시간 함께 성장할 수 있는 회사를 만들고자 한다.
- CX : 자신의 성장을 느낄 수 있는 회사, 성과에 따른 적절한 보상과 효율적인 업무를 진행할 수 있는 회사
- DX : 데이터 분석을 통해 이직률과 연관있는 사항들을 지속적으로 확인하고 개선, HR팀은 이직 가능성이 높은 핵심 인재를 예측하고 해당 인원을 위한 관리 프로그램을 운영
- Trigger : HR 데이터를 통한 최근 퇴사 및 이직 현황 확인, 핵심 인재의 유출 여부 확인
- Accelerator : 직원의 인사관련 데이터를 통해 구성원의 불만족 요소나 잠재적인 문제(진급, 보상) 확인
- Tracker : 핵심 인재의 이직률의 변화 추이, 직원들의 업무 스트레스 지수

### 탐색적 데이터 분석 

```python
# 이직 여부 
df['Attrition'].value_counts()
```

```python
# 이직률  약 16%
237/1470
```

```python
# 학습 및 집계를 하기 위해 수치 데이터로 변경
# 이직안함 = 0, 이직함 = 1로 수정
df['Attrition'] = np.where(df['Attrition'] == 'Yes', 1, 0)
df['Attrition'].value_counts()
```

```python
# 성별, 연령분석
# no.where -> 오렌지에서 fomula와 같은 역할을 하는 함수
# 구간화
# where  기능이 2개의 구간으로 나눠주는 함수
df['Age_gp'] = np.where(df['Age'] <= 30, "30이하",
                       np.where(df['Age'] <= 40, '30-40대', '40대 이상'))
```

```python
df[['Age','Age_gp']]
```

```python
# 연령대별 이직률 현황
# Age_gp 열을 기준으로 그룹화 하고, 각 그룹에 대해 특정 통계치를 계산해서 새로운 데이터프레임 만들기
# agg 함수는 aggregate의 약자로 하나 이상의 집계 함수를 적용해서 그룹별 집계 결과를 얻는 함수
df_gp = df.groupby('Age_gp')['Attrition'].agg(['count', 'sum'])
df_gp['ratio'] = round((df_gp['sum'] / df_gp['count']) *100, 1)
df_gp
# 나이가 어릴수록 이직 확률이 높다.
```

```python
# 성별 이직률 현황
df_gp = df.groupby('Gender')['Attrition'].agg(['count', 'sum'])
df_gp['ratio'] = round((df_gp['sum'] / df_gp['count']) *100, 1)
df_gp
# 남성의 이직률이 높다.
```

```python
# 부서별 이직률 현황
df_gp = df.groupby('Department')['Attrition'].agg(['count', 'sum'])
df_gp['ratio'] = round((df_gp['sum'] / df_gp['count']) *100, 1)
df_gp
# Sales > human Resources > R&D 부서순으로 이직이 많다.
```

```python
# 가설 1) 업무 만족도는 높으나, 인간관꼐로 인한 이직률이 높을 것이다.
# JobSatisfaction : 업무 만족도 컬럼 -> 숫자가 클수록 만족
# RelationshipSatisfaction : 인간관계 만족도 
df[['JobSatisfaction', 'RelationshipSatisfaction', 'Attrition']].head()
```

```python
# 업무 만족도 별 이직률 현황
df_gp = df.groupby('JobSatisfaction')['Attrition'].agg(['count', 'sum'])
df_gp['ratio'] = round((df_gp['sum'] / df_gp['count']) *100, 1)
df_gp
# 업무 만족도가 낮을수록 이직률이 높다.
```

```python
# 업무 만족도, 인관관계 만족도 별 이직률 현황
df_gp = df.groupby(['JobSatisfaction', 'RelationshipSatisfaction'])['Attrition'].agg(['count', 'sum'])
df_gp['ratio'] = round((df_gp['sum'] / df_gp['count']) *100, 1)
df_gp
# 업무 만족도가 높은 직원은 인간관계에 따라 이직률이 변하는 것이 영향이 적은 것으로 보이고,
# 업무 만족도가 낮은 직원은 인간관계가 나쁠수록 이직률이 증가하는 경향이 보인다.
```

```python
# 가설2) 근속년수 대비 같은 업무를 한 비중이 높다면 이직률이 높을 것이다.
# YearsInCurrentRole : 직원이 현재 역할에서 근무한 기간
# YearsAtCompany : 직원이 현재까지 근무한 기간
df[['YearsInCurrentRole', 'YearsAtCompany']]
```

```python
# 근속년수 대비 한가지 업종의 역할을 수행한 비중
df['Role_Company'] = df['YearsInCurrentRole'] / df['YearsAtCompany']
df['Role_Company'].fillna(0, inplace = True)
```

```python
# 데이터 분포 확인
sns.distplot(df['Role_Company'])
```

```python
# 구간화
df['Role_company_gp'] = np.where(df['Role_Company'] <= 0.3, "0.3이하",
                       np.where(df['Role_Company'] <= 0.6, '0.3~0.6', '0.6 초과'))
df[['Role_Company', 'Role_company_gp']]
```

```python
# 근속년수 대비 같은일을 오래한 경우 이직률 현황
df_gp = df.groupby(['Role_company_gp'])['Attrition'].agg(['count', 'sum'])
df_gp['ratio'] = round((df_gp['sum'] / df_gp['count']) *100, 1)
df_gp
# 근속년수 대비 같은일을 오래한 경우 이직률이 낮다
```

```python
# 가설 3) 야근을 많이해도 급여 인상률이 높다면 이직률이 낮을 것이다.
# OverTime : 연장근무 시간
df_gp = df.groupby(['OverTime'])['Attrition'].agg(['count', 'sum'])
df_gp['ratio'] = round((df_gp['sum'] / df_gp['count']) *100, 1)
df_gp
```

```python
# 'PercentSalaryHike'
df_gp = df.groupby(['OverTime', 'PercentSalaryHike'])['Attrition'].agg(['count', 'sum'])
df_gp['ratio'] = round((df_gp['sum'] / df_gp['count']) *100, 1)
df_gp
```

```python
# 야근을 하지 않는 직원 이직률 평균을 확인해보자
df_dp_plot = df_gp.reset_index()
df_dp_plot = df_dp_plot[df_dp_plot['OverTime'] == 'No']
df_dp_plot['ratio'].mean()
# 전체 이직률 16% -> 야근 No 직원은 연봉 인상률과 관계 없이 이직률이 전체 평균 이직률 보다 낮다.
```

```python
# 야근을 하지 않는 직원 이직률 평균을 확인해보자
df_dp_plot = df_gp.reset_index()
df_dp_plot = df_dp_plot[df_dp_plot['OverTime'] == 'Yes']
df_dp_plot['ratio'].mean()
# 야근을 많이하는 직원일수록 이직률이 높구나
```

### 모델 선택 및 하이퍼 파라미터 분석

```python
df
```

```python
# 데이터 타입 확인해보기
for i in df.columns :
    print(df[i].dtypes, df[i].dtypes == 'O') # 대문자 O -> Object
```

```python
# 수치형 / 카테고리 형으로 데이터를 나눠보자
numeric_list = []
categorical_list = []

for i in df.columns :
    if df[i].dtypes == 'O' : # 대문자 O -> Object
        categorical_list.append(i)
    else :
        numeric_list.append(i)

print("categorical : ", categorical_list)
print("numeric : ", numeric_list)
```

```python
# 가설 검정을 위한 컬럼들을 삭제
X = df.drop(['Attrition', 'Age_gp', 'Role_company_gp', 'Role_Company'], axis = 1)
y = df['Attrition']
```

```python
# 카테고리컬 데이터 - object 타입의 데이터를 살펴 보자
# 'BusinessTravel' : 출장 여부
# 출장이 많을 수록 이직률이 높지 않을까?
X['BusinessTravel'].unique()
```

```python
# 딕셔너리로 정의(우선순위를 주고싶은 알파벳에 큰 숫자 부여)
BT_dict = {'Travel_Rarely' : 1,
           'Travel_Frequently' : 2,
           'Non-Travel' : 0}

X['BusinessTravel'] = X['BusinessTravel'].map(BT_dict)
X['BusinessTravel']
```

```python
# 원핫 인코딩
X_one_hot = pd.get_dummies(X, dtype= 'int64')
X_one_hot
```

```python
# 정답 분할 도구 불러오기
from sklearn.model_selection import train_test_split
```

```python
# 데이터 분리 
X_train, X_test, y_train, y_test = train_test_split(X_one_hot,         # 문제 데이터
                                                   y,                  # 정답 데이터
                                                   test_size = 0.3,    # 정답 데이터의 비율
                                                   random_state = 12,  # 랜덤 고정값 지정
                                                   stratify = y        # 정답 클래스의 비율을 맞춰주세요
                                                   )
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# 모델링 시작 
from sklearn.linear_model import LogisticRegression
# 로지스틱 회귀 : 이름은 회귀지만 분류모델이다.
```

```python
logi = LogisticRegression()
```

```python
# 모델 힉습 : fit
logi.fit(X_train, y_train)
```

```python
# 예측 및 성능평가 
# 정답 예측
pred = logi.predict(X_test)
```

```python
pred
```

```python
# 정확도 살펴보기
from sklearn.metrics import accuracy_score as acc # 분류에서 정확도
acc(y_test, pred)
```

### 교차검증 실행해보기 
- 이 모델에 일반화 성능은 어느정도일까?
- 성능 평가를 진행할 때 모델이 정답을 얼마나 잘 맞추는가? 평가 -> 한번에 평가를 하는것보다 나눠서 평가를 하는것이 신뢰도가 높다

```python
# 교차검증
# 과대적합이니지 확인을 진행하는 방법 -> 직접적인 제어를 하는것은 아니다.
# 일반화 성능 정도를 파악만 하는 것 
from sklearn.model_selection import cross_val_score
print(cross_val_score(logi, X_train, y_train, cv = 5))
print(cross_val_score(logi, X_train, y_train, cv = 5).mean())
```

```python
# 로지스틱 회귀에 대한 분류평가 보고서 출력해보기
from sklearn.metrics import classification_report

print(classification_report(y_test, pred))
```

```python
# 일반화 성능을 살펴보자
print("train score :", logi.score(X_train, y_train))
print("test score :", logi.score(X_test, y_test))
```

```python
# 모델에 가장 많은 영향을 미친 변수를 탐색
importance_value = logi.coef_[0]
importance = pd.Series(importance_value, index = X_train.columns)
top10 = importance.sort_values(ascending = False)[:10]
top10
```
